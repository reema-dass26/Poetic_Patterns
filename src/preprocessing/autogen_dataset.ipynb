```python
import random
import pandas as pd

# Step 1: Define emotion categories
emotions = [
    'joy', 'sadness', 'anger', 'fear', 'surprise', 'disgust',
    'hope', 'love', 'grief', 'awe', 'yearning', 'empowerment'
]

# Step 2: Generate 30 poems per emotion using structured templates
def generate_poem(emotion):
    opening_lines = [
        f"This is a poem about {emotion}.",
        f"In this verse, {emotion} takes form.",
        f"Let us wander through the feeling of {emotion}.",
        f"A whisper of {emotion} fills the lines below."
    ]
    body_lines = [
        f"Where thoughts unravel and {emotion} flows,",
        f"And every word a mirror of {emotion}'s grace.",
        f"Wrapped in metaphors that echo {emotion},",
        f"The stanzas pulse with {emotion}'s beat."
    ]
    closing_lines = [
        f"Until the final line exhales {emotion}.",
        f"Leaving only the shadow of {emotion} behind.",
        f"A trace of {emotion} lingers in the air.",
        f"And {emotion} sleeps within the rhyme."
    ]
    poem = "\n".join([
        random.choice(opening_lines),
        random.choice(body_lines),
        random.choice(body_lines),
        random.choice(closing_lines)
    ])
    return poem

# Step 3: Create dataset
data = []
for emotion in emotions:
    for i in range(30):
        data.append({
            "title": f"Synthetic {emotion.capitalize()} Poem {i+1}",
            "poem": generate_poem(emotion),
            "emotion": emotion,
            "author": "GPT-Synth",
            "topic": emotion,
            "form": "free verse"
        })

synthetic_df = pd.DataFrame(data)
synthetic_df.to_csv("synthetic_12_emotions_dataset.csv", index=False)
synthetic_df.head()



# Save to CSV
output_path = "./data_cleaned/synthetic_12_emotions_dataset.csv"
synthetic_df.to_csv(output_path, index=False)

# import ace_tools as tools; tools.display_dataframe_to_user(name="Synthetic Poetry Dataset", dataframe=synthetic_poems_df)
# output_path
```
